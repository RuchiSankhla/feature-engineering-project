1: Data Cleaning

1.1: Handle Missing Values

```python
df.isnull().sum()
```

Fill numeric missing values with median

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

Fill categorical missing values with mode

```python
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])
```

1.4: Outlier Detection & Treatment

```python
import numpy as np

Q1 = df['Monthly_Income'].quantile(0.25)
Q3 = df['Monthly_Income'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['Monthly_Income'] = np.where(
    df['Monthly_Income'] > upper, upper,
    np.where(df['Monthly_Income'] < lower, lower, df['Monthly_Income'])
)
```

2: Data Integration is done earlier. 

2.1: Standardize Column Names

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.columns
```

2.2: Check for Multiple Forms of IDs / Names

```python
if 'emp_id' in df.columns and 'employee_id' in df.columns:
    df['employee_id'] = df['employee_id'].fillna(df['emp_id'])
    df.drop(columns=['emp_id'], inplace=True)
```

2.3: Entity Standardization:
In real-world HR datasets, employee names may appear in multiple formats (e.g., “J Sharma” vs “Jyoti Sharma”). We normally use string cleaning and fuzzy matching to standardize such values. Since our dataset uses a unique Employee_ID field and does not contain names, no entity correction was required.

2.4: Unit Standardization:
In real-world data integration, salary or time units may differ across files (e.g., USD vs INR). A conversion step is required to maintain consistency.
Our dataset already uses a uniform currency and time units, so no conversion was needed.

3: Data Transformation

3.1: Encode Categorical Data

```python
from sklearn.preprocessing import LabelEncoder

label_enc = LabelEncoder()

binary_cols = ['gender', 'overtime', 'attrition']

for col in binary_cols:
    df[col] = label_enc.fit_transform(df[col])
```

```python
df[['gender', 'overtime', 'attrition']].head()
```

3.2: Normalize / Standardize Numerical Data

Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
num_cols = df.select_dtypes(include=['int64','float64']).columns

df[num_cols] = scaler.fit_transform(df[num_cols])
```

3.3: Log Transform Skewed Columns

```python
import numpy as np

skewed_cols = ['monthly_income','distance_from_home','training_hours_last_year']

for col in skewed_cols:
    df[col] = np.log1p(df[col])
```
